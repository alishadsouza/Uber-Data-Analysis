```python
import numpy as np
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv('/content/UberDataset.csv')

# size of the dataset
df.shape
df.head()
```

```python
# Let's explore variables, their data types, and total non-null values
df.info()
```

```python
# Filling NULL values in feature PURPOSE
df['PURPOSE'] = df['PURPOSE'].fillna("UNKNOWN")

# After this imputation, we are left with 1 NULL value in features - END_DATE, CATEGORY, START, and STOP. We can handle the NULL values by dropping that record
df.dropna(inplace = True)

df.isnull().sum()
```

```python
# converting START_DATE and END_DATE into datetime format
df['START_DATE'] = pd.to_datetime(df['START_DATE'])
df['END_DATE'] = pd.to_datetime(df['END_DATE'])

#create new feature representing time of the ride - morning, afternoon, evening, night
df['TIME_DAY'] = df['START_DATE'].apply(lambda i : i.hour)
df['TIME_OF_DAY'] = df['START_DATE'].apply(lambda i : i.hour)
df['TIME_OF_DAY'] = pd.cut(x = df['TIME_OF_DAY'], bins = [0, 6, 11, 17, 21, 24],\
                             labels = ['Night', 'Morning','Afternoon','Evening','Night'], ordered = False)

#create new feature representing month of the ride
df['MONTH_OF_THE_RIDE'] = df['START_DATE'].apply(lambda i : i.month)
month_label = {1.0: 'Jan', 2.0: 'Feb', 3.0: 'Mar', 4.0: 'April', 5.0: 'May', 6.0: 'June', 7.0: 'July', 8.0: 'Aug', 9.0: 'Sep', 10.0: 'Oct', 11.0: 'Nov', 12.0: 'Dec'}
df['MONTH_OF_THE_RIDE'] = df['MONTH_OF_THE_RIDE'].map(month_label)
#create new feature representing day (week) of the ride
df['DAY_OF_THE_RIDE'] = df['START_DATE'].apply(lambda i : i.weekday())
day_label = {0: 'Mon', 1: 'Tues', 2: 'Wed', 3: 'Thus', 4: 'Fri', 5: 'Sat', 6: 'Sun'}

df['DAY_OF_THE_RIDE'] = df['DAY_OF_THE_RIDE'].map(day_label)

# duration of the ride
df['DURATION_OF_THE_RIDE'] = (df['END_DATE'] - df['START_DATE']).astype('timedelta64[m]')

df.head()
```

```python
# summary statistics of the dataset
df.describe()
```

```python
print('Min Timestamp in START_TIME is - {}'.format(df['START_DATE'].min()))
print('Max Timestamp in START_TIME is - {}'.format(df['START_DATE'].max()))
```

```python
print("Total number of unique categories in CATEGORY - {}".format(df['CATEGORY'].nunique()))
print("Total number of unique categories in PURPOSE - {}".format(df['PURPOSE'].nunique()))
print("Total number of unique location in START - {}".format(df['START'].nunique()))
print("Total number of unique location in STOP - {}".format(df['STOP'].nunique()))
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(18,18))

# label distribution for CATEGORY
df['CATEGORY'].value_counts().plot(kind='bar', ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Categories in CATEGORY')

# label distribution for PURPOSE
df['PURPOSE'].value_counts().plot(kind='bar', ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Categories in PURPOSE')
axes[1, 0].tick_params(axis='x', rotation=90)

# top 10 locations for START and STOP
temp = df['START'].value_counts().nlargest(10)
temp.plot(kind='bar', ax=axes[0, 1])
axes[0, 1].set_title('Top 10 START Locations')
axes[0, 1].tick_params(axis='x', rotation=90)

temp = df['STOP'].value_counts().nlargest(10)
temp.plot(kind='bar', ax=axes[1, 1])
axes[1, 1].set_title('Top 10 STOP Locations')
axes[1, 1].tick_params(axis='x', rotation=90)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Group the data by 'TIME_OF_DAY' and count the number of rides in each category
ride_counts = df.groupby('TIME_OF_DAY').size().reset_index(name='Ride Count')

# Create a bar plot
sns.barplot(x='TIME_OF_DAY', y='Ride Count', data=ride_counts)

plt.xlabel('Time of Day')
plt.ylabel('Number of Rides')
plt.title('Distribution of Rides Based on Time of the Day')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Group the data by 'MONTH_OF_THE_RIDE' and count the number of rides in each category
ride_counts = df.groupby('MONTH_OF_THE_RIDE').size().reset_index(name='Ride Count')

# Create a bar plot
sns.barplot(x='MONTH_OF_THE_RIDE', y='Ride Count', data=ride_counts)

plt.xlabel('Month of the Ride')
plt.ylabel('Number of Rides')
plt.title('Distribution of Rides Based on Month of the Ride')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Calculate the counts for each unique value in 'DAY_OF_THE_RIDE'
ride_counts = df['DAY_OF_THE_RIDE'].value_counts()

# Create a bar plot
ride_counts.plot(kind='bar')

plt.xlabel('Day of the Ride')
plt.ylabel('Number of Rides')
plt.title('Distribution of Rides Based on Day of the Ride')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Calculate the average duration of rides for each month
temp = df.groupby('MONTH_OF_THE_RIDE')['DURATION_OF_THE_RIDE'].mean().reset_index()

# Create a bar plot
plt.bar(temp['MONTH_OF_THE_RIDE'], temp['DURATION_OF_THE_RIDE'])

plt.xlabel('Month of the Ride')
plt.ylabel('Average Duration')
plt.title('Average Duration of Rides by Month')

plt.show()
```

```python
# Average duration of rides based on time of the day
plt.figure(figsize=(12, 6))
sns.barplot(x='TIME_OF_DAY', y='DURATION_OF_THE_RIDE', data=df)
plt.xlabel('Time of the Day')
plt.ylabel('Average Duration')
plt.title('Average Duration of Rides by Time of the Day')
plt.show()

# Average miles of the rides based on time of the day
plt.figure(figsize=(12, 6))
sns.barplot(x='TIME_OF_DAY', y='MILES', data=df)
plt.xlabel('Time of the Day')
plt.ylabel('Average Miles')
plt.title('Average Miles of Rides by Time of the Day')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['DURATION_OF_THE_RIDE'], df['MILES'])
plt.xlabel('Duration')
plt.ylabel('Miles')
plt.title('Scatter Plot of Duration vs. Miles')
plt.show()
```

```python
# Subset the data
subset_df = df.sample(n=1000, random_state=42)  # Select a random sample of 1000 rows

# Select a subset of variables
selected_vars = ['DURATION_OF_THE_RIDE', 'MILES', 'CATEGORY', 'PURPOSE']
subset_df = subset_df.loc[:, selected_vars]


# One-hot encode categorical features
encoded_df = pd.get_dummies(subset_df, columns=['CATEGORY', 'PURPOSE'])

# Calculate correlation matrix
correlation_matrix = encoded_df.corr()

# Create heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
